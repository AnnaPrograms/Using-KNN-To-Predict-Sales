## Using KNN To Predict Sales

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsRegressor
from sklearn.model_selection import train_test_split
%matplotlib inline
```

### Setting up KNN by hand

```python
df = pd.read_csv("data/Advertising.csv", usecols = ["TV", "Sales"])
```

```python
# we want a subset of the data 5 to 40
x_true = df.TV.iloc[5:40]
y_true = df.Sales.iloc[5:40]

# sort the arry to get the indexs by lowest to highest
idx = np.argsort(x_true).values

# get the predictor data in the order by idx
x_true = x_true.iloc[idx]

# get the response data in the order by idx
y_true = y_true.iloc[idx]
```

```python
def calculate_distance(x1, x2): 
    distance = np.sqrt(np.sum(x1-x2)**2)
    return distance

def find_nearest(array, value):
    idx = pd.DataFrame(calculate_distance(array, value)).idxmin()
    return idx, array[idx]
```

```python
# create some synthetic x-values (might not be in the actual dataset)
x = np.linspace(np.min(x_true), np.max(x_true))

# initialize the y-values for the length of the synthetic x-values to zero
y = np.zeros(x.shape)
```

```python

# apply the KNN algorithm to predict the y-value for the given x value
for i, xi in enumerate(x):
    
    # get the Sales values closest to the given x value
    y[i] = y[find_nearest(x, xi)[0]]
    
```

### Plot Our Graphs

```python

# Plot the synthetic data along with the predictions    
plt.plot(x, y, c = "g", linestyle = "--")

# plot original
plt.plot(x_true, y_true, c= "y")
plt.scatter(x_true, y_true)
```
